```python
import pandas as pd
import folium
df = pd.read_excel("/content/자연휴양림출력파일.xlsx")
df.head()
```

```python
anMap = folium.Map(location=[37.4965,126.9573],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale=True)
anMap
```

```python
for n, row in df.iterrows():
  mk = folium.Marker(location = [row['위도'],row['경도']], popup = folium.Popup(row['휴양림명'],max_width=300))
  mk.add_to(anMap)
anMap
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap)

for n, row in df.iterrows():
  folium.Marker(location = [row['위도'],row['경도']], popup = folium.Popup(row['휴양림명'],max_width=300)).add_to(mc)
anMap
```

```python
anMap.save("anMap.html")
```
